# CRMS: Cloth Rental Management System

**Importing necessary modules**

```python
from tkinter import *
from tkinter import messagebox     # for asking user Quit as Yes/No and for some info display
import mysql.connector             # Simply for storing cust details in MySQL
from tkinter import ttk            # ttk is 'Themed Tk' for better appearance. Used for ttk.Combobox() which is a 'dropdown'
```

**Main class**

```python
class CRMS:

    def __init__(self, root):     # initialising by Constructor
        self.root=root
        self.root.title("Cloth Rental Management System")
        self.root.geometry("1366x768+100+0")     # This additional params(+100+0) decides the position of Tkinter window
        self.root.config(bg="#6600cc")

        
        '''
        Similar to Entry widget, StringVar class of tkinter is use to set as well get variables
        '''
        customerId = StringVar()        
        idc = StringVar()
        mtype = StringVar()
        clothAmount = StringVar()
        clothSize = StringVar()
        deposit = StringVar()
        firstName = StringVar()
        dateBorrowed = StringVar()
        lastName = StringVar()
        dueDate = StringVar()
        gender = StringVar()
        dateOverDue = StringVar()

        #================================Functions===========================
        def Exit():
            qExit= messagebox.askyesno("Quit System","Do you want to Quit?")
            if qExit :
                root.destroy()      # destroys the 'root' window
                return

        def clearData():
            self.txtmtype.delete(0, END)    # remove items or ranges of items from certain widgets
            self.txtidc.delete(0, END)
            self.txtcustomerId.delete(0, END)
            self.txtclothAmount.delete(0, END)
            clothSize.set("")
            self.txtdeposit.delete(0, END)
            self.txtfirstName.delete(0, END)
            self.txtdateBorrowed.delete(0, END)
            self.txtlastName.delete(0, END)
            self.txtdueDate.delete(0, END)
            gender.set("")
            self.txtdateOverDue.delete(0, END)

        def addData():
            if(mtype.get()=="" or idc.get()=="" or dueDate.get()=="" or customerId.get()==""or clothAmount.get()=="" or clothSize.get()=="" or deposit.get()=="" or firstName.get()=="" or dateBorrowed.get()=="" or lastName.get()=="" or gender.get()=="" or dateOverDue.get()==""):
                messagebox.showinfo("Insert Status","All Fields are Compulsary!")
                return False   # returning to update()
            else:
                mydb = mysql.connector.connect(host="localhost",user="admin",passwd="admin123",database="CRMS")
                mycursor = mydb.cursor()
                mycursor.execute("INSERT into clothes values('"+str(customerId.get())+"','"+str(idc.get())+"','"+str(mtype.get())+"','"+str(clothAmount.get())+"','"+str(clothSize.get())+"','"+str(deposit.get())+"','"+str(firstName.get())+"','"+str(lastName.get())+"', '"+str(gender.get())+"', '"+str(dateBorrowed.get())+"','"+str(dateOverDue.get())+"', '"+str(dueDate.get())+"')")
                mycursor.execute("commit")
                messagebox.showinfo("Insert Status","Inserted Successfully!")
                mydb.close()    
                return True   

        def displayData():
            clothlist.delete(0,END)
            mydb = mysql.connector.connect(host="localhost",user="admin",passwd="admin123",database="CRMS")
            mycursor = mydb.cursor()
            mycursor.execute("SELECT * from clothes")
            rows = mycursor.fetchall()
            for row in rows:
                clothlist.insert(END, row, str(""))
            mydb.close()
            return rows          
            
        def deleteData():
            if(customerId.get()==""):
                messagebox.showinfo("Insert Status","Customer Id is Compulsary!")
            else:
                mydb = mysql.connector.connect(host="localhost",user="admin",passwd="admin123",database="CRMS")
                mycursor = mydb.cursor()
                mycursor.execute("DELETE FROM clothes WHERE customerId='"+customerId.get()+"'")
                mycursor.execute("commit")
                messagebox.showinfo("Delete Status","Deleted Successfully!")
                displayData()                
                mydb.close()

        def update():
            if(customerId.get()==""):
                messagebox.showinfo("Insert Status","Customer Id is Compulsary!")
            else:
                mydb = mysql.connector.connect(host="localhost",user="admin",passwd="admin123",database="CRMS")
                mycursor = mydb.cursor()
                mycursor.execute("DELETE FROM clothes WHERE customerId='"+customerId.get()+"'")
                mycursor.execute("commit")
                
                if addData():      # show successful status only if all fields are given 
                    messagebox.showinfo("Update Status","Updated Successfully!")
                mydb.close()           
        
        #==================================Frames===================================
        
        '''
        Frame widget/class is used to group and organize other widgets like Buttons, Labels,Entry fiels,etc and gives a 3-D border
        '''
                        # ------------      Frame-1: Main        ------------ #

        MainFrame=Frame(self.root)     # Main frame for Container        
        MainFrame.grid()
                        # ------------      Frame-2: Header        ------------ #

        ProjectNameFrame=Frame(MainFrame, bd=17, padx=40, pady=8, bg="#6600cc", relief=SUNKEN)     # Header name Frame; relief specifies visual relief (border style) of Frames
        ProjectNameFrame.pack(side=TOP)

        self.lblcustomerId=Label(ProjectNameFrame, font=('arial', 46, 'bold'), text="Cloth Rental Management System")
        self.lblcustomerId.grid()
        
                        # ------------      Frame-3: Button Frame        ------------ #

        ButtonFrame=Frame(MainFrame, bd=2, width=1050, height=100, padx=20, pady=20, bg="#6600cc", relief=SUNKEN)    
        ButtonFrame.pack(side=BOTTOM)      # Buttons will be alligned at bottom

                        # ------------      Frame-4: DataFrame        ------------ #

        DataFrame=Frame(MainFrame, bd=1, width=1000, height=400, padx=20, pady=20, relief=SUNKEN)   # Data Frame for Labels and Entry Field
        DataFrame.pack(side=BOTTOM)      

        DataFrameLeft=LabelFrame(DataFrame, bd=1, width=800, height=300, padx=20, relief=SUNKEN, font=('arial', 12, 'bold'), text="Customer Info:", bg="#6600cc")
        DataFrameLeft.pack(side=LEFT)       # DataFrameLeft for Lables and Entry

        DataFrameRight=LabelFrame(DataFrame, bd=1, width=450, height=300, padx=20, pady=3, relief=SUNKEN, font=('arial', 12, 'bold'), bg="#6600cc", text="Details:")
        DataFrameRight.pack(side=RIGHT)      # DataFrameRight for Listbox and  Scrollbar

        #===============================Label and Entry==================================
        
        self.lblmtype=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Customer ID:", padx=2, pady=2, bg="#6600cc")
        self.lblmtype.grid(row=0, column=0, sticky=W)
        self.txtmtype=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=customerId, width=20)
        self.txtmtype.grid(row=0, column=1)

        self.lblidc=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Cloth ID:", padx=2, pady=2, bg="#6600cc")
        self.lblidc.grid(row=0, column=2, sticky=W)
        self.txtidc=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=idc, width=20)
        self.txtidc.grid(row=0, column=3)

        self.lblcustomerId=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Member Type:", padx=2, pady=2, bg="#6600cc")
        self.lblcustomerId.grid(row=1, column=0, sticky=W)
        self.txtcustomerId=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=mtype, width=20)
        self.txtcustomerId.grid(row=1, column=1)

        self.lblclothAmount=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Cloth Amount:", padx=2, pady=2, bg="#6600cc")
        self.lblclothAmount.grid(row=1, column=2, sticky=W)
        self.txtclothAmount=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=clothAmount, width=20)
        self.txtclothAmount.grid(row=1, column=3)

        self.lblclothSize=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Cloth Size:", padx=2, pady=2, bg="#6600cc")
        self.lblclothSize.grid(row=2, column=0, sticky=W)
        self.dropdown=ttk.Combobox(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=clothSize, width=18)
        self.dropdown['value']=('', 'S', 'M','L','XL','XXL')
        self.dropdown.current(0)
        self.dropdown.grid(row=2, column=1)

        self.lbldeposit=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Deposit:", padx=2, pady=2, bg="#6600cc")
        self.lbldeposit.grid(row=2, column=2, sticky=W)
        self.txtdeposit=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=deposit, width=20)
        self.txtdeposit.grid(row=2, column=3)

        self.lblfirstName=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="First Name:", padx=2, pady=2, bg="#6600cc")
        self.lblfirstName.grid(row=3, column=0, sticky=W)
        self.txtfirstName=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=firstName, width=20)
        self.txtfirstName.grid(row=3, column=1)

        self.lbldateBorrowed=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Date Borrowed:", padx=2, pady=2, bg="#6600cc")
        self.lbldateBorrowed.grid(row=3, column=2, sticky=W)
        self.txtdateBorrowed=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=dateBorrowed, width=20)
        self.txtdateBorrowed.grid(row=3, column=3)

        self.lbllastName=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Last Name:", padx=2, pady=2, bg="#6600cc")
        self.lbllastName.grid(row=4, column=0, sticky=W)
        self.txtlastName=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=lastName, width=20)
        self.txtlastName.grid(row=4, column=1)

        self.lbldueDate=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Due Date:", padx=2, pady=2, bg="#6600cc")
        self.lbldueDate.grid(row=4, column=2, sticky=W)
        self.txtdueDate=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=dueDate, width=20)
        self.txtdueDate.grid(row=4, column=3)

        self.lblgender=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Gender:", padx=2, pady=2, bg="#6600cc")
        self.lblgender.grid(row=5, column=0, sticky=W)
        self.dropdown=ttk.Combobox(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=gender, width=18)
        self.dropdown['value']=('', 'Male', 'Female','Other')
        self.dropdown.current(0)
        self.dropdown.grid(row=5, column=1)
        
        self.lbldateOverDue=Label(DataFrameLeft, font=('arial', 12, 'bold'), text="Date Over Due:", padx=2, pady=2, bg="#6600cc")
        self.lbldateOverDue.grid(row=5, column=2, sticky=W)
        self.txtdateOverDue=Entry(DataFrameLeft, font=('arial', 12, 'bold'), textvariable=dateOverDue, width=20)
        self.txtdateOverDue.grid(row=5, column=3)

        #================================Listbox and Scrollbar================

        scrollbar=Scrollbar(DataFrameRight)
        scrollbar.grid(row=0, column=1, sticky='ns')     # will stick the scrollbar as North-South (Top-Bottom)

        clothlist=Listbox(DataFrameRight, width=45, height=12, font=('arial', 12, 'bold'), yscrollcommand=scrollbar.set)
        clothlist.grid(row=0, column=0, padx=8)
        scrollbar.config(command=clothlist.yview)

        #====================Buttons==========================================
        
        self.btnadddata=Button(ButtonFrame, text='Add Data', font=('arial', 12, 'bold'), height=2, width=10, bd=4, command=addData)
        self.btnadddata.grid(row=0, column=0)

        self.btndisplaydata=Button(ButtonFrame, text='Display Data', font=('arial', 12, 'bold'), height=2, width=10, bd=4, command=displayData)
        self.btndisplaydata.grid(row=0, column=1)

        self.btncleardata=Button(ButtonFrame, text='Clear Data', font=('arial', 12, 'bold'), height=2, width=10, bd=4, command=clearData)
        self.btncleardata.grid(row=0, column=2)

        self.btndeletedata=Button(ButtonFrame, text='Delete Data', font=('arial', 12, 'bold'), height=2, width=10, bd=4, command=deleteData)
        self.btndeletedata.grid(row=0, column=3)

        self.btnupdatedata=Button(ButtonFrame, text='Update Data', font=('arial', 12, 'bold'), height=2, width=10, bd=4, command=update)
        self.btnupdatedata.grid(row=0, column=4)

        self.btnexit=Button(ButtonFrame, text='Exit', font=('arial', 12, 'bold'), height=2, width=10, bd=4, command=Exit)
        self.btnexit.grid(row=0, column=5)
        
if __name__=='__main__':
    root=Tk()   # root is our main waindow (a.k.a Container)
    '''
     When instantiated 'Tk' class, creates a top-level window which acts as the main container for other GUI elements, 
     such as buttons, labels, entry fields,etc.
    '''
    CRMS(root)     # Instantiating CRMS class
    root.mainloop()
    
    '''
    mainloop() method starts the event loop. The event loop is responsible for detecting user interactions 
    such as button clicks or key presses, and updating the GUI accordingly. It also blocks any code that comes after 
    it from running until you close the window where you called the method.
    '''
    
    
```
